```python
import yfinance as yf 
import requests
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0 Safari/537.36'}
html=requests.get(url,headers=headers) 
sp500= pd.read_html(html.text)[0] 
sp500 


```

```python
end_date= '2025-10-27' 
start= pd.to_datetime(end_date)-pd.DateOffset(365*8)  
df= yf.download(tickers='TSLA', start=start, end=end_date).stack()
df
```

```python
 df['Open'].plot(figsize=(15,6))
```

```python
df['open:10 day rolling']= df['Open'].rolling(window=10).mean() 
df['open:20 day rolling']= df['Open'].rolling(window=20).mean() 
df['open:50 day rolling']= df['Open'].rolling(window=50).mean()  
df.head()
```

```python
df[['Open','open:10 day rolling','open:20 day rolling','open:50 day rolling']].plot(figsize=(15,6))
```

```python
df.loc['2017-10-30':'2025-10-21'].mean()['Close'] 
```

```python
df_tesla=df.reset_index()  
df_tesla.head()
df_tesla.set_index('Date',drop=True)
```

```python
df_tesla.resample(rule='A', on='Date').min()
```

```python
df_tesla.resample(rule='B',on='Date').max()['Open'].plot()
```

```python
df_tesla.resample(rule='QS',on='Date').max()['Close'].plot()
```

```python
df_tesla.columns
```

```python
df_tesla.index
```

```python
# Resample the DataFrame on the 'Date' column, then select 'Open' and get the mean
df_tesla.resample(rule='BA', on='Date')['Open'].mean().plot(kind='bar', figsize=(15,6))
```

```python
df_tesla['Close'].expanding().mean().plot(figsize=(15,6))
```

```python
df_tesla['EWMA-0.1']=df_tesla['Close'].ewm(alpha=0.1,adjust=False).mean() 

df_tesla['EWMA-0.2']=df_tesla['Close'].ewm(alpha=0.2,adjust=False).mean() 
df_tesla['EWMA-0.3']=df_tesla['Close'].ewm(alpha=0.3,adjust=False).mean() 
df_tesla.head(10)

```

```python
df_tesla['EWMA-0.1'].plot(figsize=(15,6))
```

```python
df_tesla[['Open','EWMA-0.1','EWMA-0.2','EWMA-0.3']].plot(figsize=(15,6))
```

```python
df_tesla['EWMA-5']=df_tesla['Close'].ewm(span=5,adjust=False).mean()
```
